```python
import numpy as np
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from PCA import unit
from PIL import Image
```

```python
tool = unit()
size = 256 * 256
k = 64
n = 3
for i in range(60):
    infile = "GSI/" + str(i) + ".jpg"
    outfile = "generalization/" + str(i) + ".jpg"
    inimg = Image.open(infile)
    data = np.mat(inimg).astype('int64')
#     averagee = np.average(data)
#     data = data - average
    pcaData = tool.pca(data, size, k, n)
#     pcaData = pcaData + average
    pcaData = np.dot(pcaData,pcaData.T)
    outimg = Image.fromarray(pcaData.astype('uint8'))
    outimg.save(outfile)
    print(pcaData)
```
